# 2015년 Data, excel 파일 만들기(64개)

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

key = 'G%2FlzSrI%2F6NyEajqlKXmhDeYc7Uiajqw3KjG8P%2Be7RTaWUDv0ZLsFLIfm%2F6GfqDlI%2F4ZOUNc3g5gIuYbDiaR1AA%3D%3D'
params = '&pageNo=1&numOfRows=64&searchYear=2015'
gov_api = 'http://apis.data.go.kr/B490007/worldjob16/openApi16?&serviceKey='+key+params

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['A'].width = 10
excel_sheet.column_dimensions['B'].width = 16
excel_sheet.column_dimensions['C'].width = 10
excel_sheet.append(['Year', 'Country', 'Count'])

res = requests.get(gov_api)
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find_all('item')
for item in mydata:
    Year = item.find('joyear')
    Country = item.find('prcerepmtnationnm')
    Count = item.find('cnt')
    excel_sheet.append([Year.get_text(), Country.get_text(), Count.get_text()])

excel_file.save('OE_2015.xlsx')
excel_file.close()
```

# 가져올 모든 dataframe에 적용시키기

```python
def create_dataframe(filename):
    data = pd.read_excel('overseas-employment/' + filename) # read a excel file
                                                            # filename: e.g., OE_2016.xlsx
    data = data[['Country','Count']]
    data = data.set_index('Country')
    
    # Change column name: 'Count' -> 'Year of File name'
    column_year = filename.split('.')[0].split('_')[1]
    data.columns = [column_year]
    return data    
```

```python
data1 = create_dataframe('OE_2015.xlsx')
data2 = create_dataframe('OE_2019.xlsx')
data2.head() # dataframe이 생성되는 것을 볼 수 있다.
```

- merge를 이용해서 두 dataframe 합치기

```python
merge_data = pd.merge(data1, data2, how='outer', left_index=True, right_index=True)
merge_data
```

- 결측치 0으로 바꾸기

```python
merge_data = merge_data.fillna(0)
merge_data
```

# 가져올 모든 dataframe에 적용시키기
- overseas-employment에 있는 모든 excel 파일 이름 출력하기

```python
import os

item_lists = os.listdir('overseas-employment')
excel_list = list()

for item in item_lists:
    if item.split('.')[-1] == 'xlsx':
        excel_list.append(item)
        
print(excel_list)
```

```python
excel_list.sort()
excel_list
```

```python
def create_dataframe_by_path(PATH):
    
    empty_df = True

    for item in excel_list:
        data = create_dataframe(item)
        if empty_df:
            empty_df = False
            final_data = data
        else:
            final_data = pd.merge(final_data, data, how = 'outer', left_index=True, right_index=True)
        
    final_data = final_data.fillna(0)
    return final_data
```

```python
PATH = 'overseas-employment/'
data = create_dataframe_by_path(PATH)
data
```

- datatype 바꾸기

```python
data = data.astype('int64')
data
```

```python
data.to_excel('final_overseas_employment.xlsx')
```

# 해외취업자 통계 excel 파일과 국가 이미지 파일의 국가 이름들을 통일시켜주기

```python
def func(df_arg):
    if df_arg['Country'] in country_name:
        df_arg['Country'] = country_name[df_arg['Country']]
    return df_arg
```
